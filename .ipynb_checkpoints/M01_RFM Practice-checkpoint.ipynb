```python
import pandas as pd
df = pd.read_csv('M01_sample_order_RFM.csv.csv')
df.head()
```

```python
df = df.iloc[:, 1:]
df.head()
```

```python
df.Date = pd.to_datetime(df.Date)
df.head()
```

```python
import datetime as dt
NOW = dt.datetime(2014,12,31)
```

```python
rfm = df.groupby('Name').agg({'Date':lambda x: (NOW-x.max()).days, 
                             'Name': lambda x: len(x), 
                             'Fare': lambda x: x.sum()})
rfm.Date = rfm.Date.astype(int)
rfm.head()
```

```python
cols = ['recency', 'frequency', 'monetory']
rfm.columns = cols
rfm.head()
```

```python
#creating the Thresold matrix 

thresholds = pd.DataFrame({'recency':[10,30,50],  'frequency': [4,6,10], 'monetory':[1000, 1500, 3000]}, index = [1,2,3])
thresholds
                          
```

```python
def RClass(x, p, d):
    if x<=d[p][1]:
        return 1
    if x<=d[p][2]:
        return 2
    if x<=d[p][3]:
        return 3
    else:
        return 4

def FMClass(x, p, d):
    if x<=d[p][3]:
        return 4
    if x<= d[p][2]:
        return 3
    if x<=d[p][1]:
        return 2
    else:
        return 1
    
    
```

```python
rfm['Recency_Q'] = rfm['recency'].apply(RClass, args=('recency', thresholds, ))
rfm['Frequency_Q'] = rfm['frequency'].apply(FMClass, args=('frequency', thresholds, ))
rfm['Monetory_Q'] = rfm['monetory'].apply(FMClass, args=('monetory', thresholds, ))
rfm['RFM_SCORE'] = rfm['Recency_Q'].map(str)+rfm['Frequency_Q'].map(str) + rfm['Monetory_Q'].map(str)
rfm.head()
```

```python
rfm[rfm['RFM_SCORE']=='111']
```

```python
len(rfm[rfm['RFM_SCORE']=='111'])
```

```python
score = []
length =[]
k = rfm['RFM_SCORE'].unique()
for i in k:
    score.append(i)
    length.append(len(rfm[rfm['RFM_SCORE']==i]))
rfm_score = pd.DataFrame(data = {'rfm_score':score, 'No. of Customers': length})
rfm_score = rfm_score[['rfm_score','No. of Customers' ]].sort_values(by='rfm_score', ascending = True)
rfm_score
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

rfm_score.plot(kind='bar', x='rfm_score', y = 'No. of Customers')
```

```python
rfm_score = rfm_score.sort_values(by='No. of Customers', ascending = True)
rfm_score.plot(kind='bar', x='rfm_score', y = 'No. of Customers')
```
